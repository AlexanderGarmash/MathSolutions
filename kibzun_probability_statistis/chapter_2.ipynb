<b>17.</b><br>
$X \sim \Pi(M[Y])$<br>
$Y \sim \Pi(M[X])$<br>
Сравнить $D[X]$ и $D[Y]$<br>
<b>Решение</b><br>
По опр. распределения пусассона имеет вид:
$\Pi(\lambda).$<br>
Значит, $X \sim \Pi(\lambda_X)$, и $Y \sim \Pi(\lambda_Y)$<br>
По свойству П, имеем:<br>
$M[X] = \lambda_X$, $M[Y] = \lambda_Y$, а по условию, имеем<br>
$\lambda_X = M[Y]$, $\lambda_Y = M[X]$, т.е. $\lambda_X = \lambda_Y = \lambda$<br>
Тогда, $M[X] = M[Y] = \lambda$, значит по свойству П, имеет<br>
$D[X] = \lambda$ и $D[Y] = \lambda$, значит $D[X] = D[Y]$

<b>Ответ: </b>$D[X] = D[Y]$
